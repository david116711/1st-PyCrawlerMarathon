## 1. 取出今天所有的新聞

```python
import requests
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
from datetime import datetime

#datetime_object = datetime.strptime(str(datetime.now())[0:19], '%Y-%m-%d %I:%M:%S').date()  
#myYear = str(datetime_object.year)
#myMonth = str(datetime_object.month)
#myDay = str(datetime_object.day)

now = datetime.now()
myYear = now.strftime('%Y')
myMonth = str(int(now.strftime('%m')))
myDay = str(int(now.strftime('%d')))

#print(myYear, myMonth, myDay)

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

selectYear = Select(browser.find_element_by_id("selY"))
selectYear.select_by_value(myYear)
selectMonth = Select(browser.find_element_by_id("selM"))
selectMonth.select_by_value(myMonth)
selectDay = Select(browser.find_element_by_id("selD"))
selectDay.select_by_value(myDay)

browser.find_element_by_id('button').click()

import time
time.sleep(120)
```

```python
html_source = browser.page_source
browser.quit();
soup = BeautifulSoup(html_source, "html5lib")
```

```python
#for d in soup.find(class_="part_list_2").find_all('h3'):
#    if d.find(class_="date").text[0:10] == '2020/01/25':
#        print(d.find(class_="date").text[0:10], d.find_all('a')[-1].text)

nowDate = now.strftime('%Y-%m-%d')   #今日日期

for d in soup.find(class_="part_list_2").find_all('h3'):
    myDate_ = datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M').date()
    myDate = myDate_.strftime('%Y-%m-%d')   #新聞日期
    if myDate == nowDate:
        print(d.find(class_="date").text[0:10], d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
from datetime import timedelta

in_hour = timedelta(hours=2)    #設定2小時差距
endingTime = now - in_hour      #2小時以內時間
#print(endingTime.strftime('%Y/%m/%d %H:%M:%S'))

for d in soup.find(class_="part_list_2").find_all('h3'):
    
    myTime = datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M')
    if myTime > endingTime:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
        
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
import re
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys

def browser_pagedown(_actions):
   time.sleep(2)
   _actions.send_keys(Keys.PAGE_DOWN)
   _actions.perform()



in_day = timedelta(days=3)         #設定3天差距
endingDateTime = now - in_day      #3天前

myYear = now.strftime('%Y')
myMonth = str(int(endingDateTime.strftime('%m')))
myDay = str(int(endingDateTime.strftime('%d')))
#print(myYear, myMonth, myDay)

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

selectYear = Select(browser.find_element_by_id("selY"))
selectYear.select_by_value(myYear)
selectMonth = Select(browser.find_element_by_id("selM"))
selectMonth.select_by_value(myMonth)
selectDay = Select(browser.find_element_by_id("selD"))
selectDay.select_by_value(myDay)

browser.find_element_by_id('button').click()

actions = ActionChains(browser)

import time
for i in range(20):
    browser_pagedown(actions)
    
    #time.sleep(3)
    #browser.execute_script("window.scrollTo(0, 10000);")
    
time.sleep(60)
    
html_source = browser.page_source
browser.quit();
soup = BeautifulSoup(html_source, "html5lib")
```

```python
startingTime = datetime(int(myYear),int(myMonth),int(myDay),15,0)
endingTime = datetime(int(myYear),int(myMonth),int(myDay),17,0)
#print(startingTime)
#print(endingTime)

for d in soup.find(class_="part_list_2").find_all('h3'):
    
    myTime = datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M')
    #print(myTime)
    if myTime > startingTime:
        if myTime < endingTime:
            print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
